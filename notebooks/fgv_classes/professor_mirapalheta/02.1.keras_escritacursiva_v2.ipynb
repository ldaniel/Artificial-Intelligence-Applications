# Representação de Dados em Uma Rede Neural

Os dados em uma rede neural são representados por tensores

#### Atributos de um tensor


Um tensor é definido:

1.   Pelo seu número de eixos. 
2.   Pelo seu formato.
3.   Pelo tipo de dado armazenado.

#### Escalares

O tensor a seguir tem 0 eixos, não tem formato e armazena um número do tipo float32. Este tensor é chamado de "escalar".

Um exemplo seria a renda anual de um cliente, 'cliente1', a qual poderia ser 65.000

```python
import numpy as np
cliente1 = np.array(65000, dtype='float32')
a, a.ndim, a.shape
```

#### Vetores

O tipo mais comum são os vetores ou tensores 1D. Neste caso temos (amostras) de uma única variável.

Por exemplo para vários clientes, registramos a renda de cada um.

O tensor a seguir tem 1 eixo, seu formato é 5 e ele armazena 5 números do tipo float32. Este tensor é chamado de "vetor"

```python
import numpy as np
a = np.array([1,2,3,4,5], dtype='float32')
a, a.ndim, a.shape
```

#### Matrizes

O tipo seguinte são as matrizes ou tensores 2D.
Neste caso temos (amostras, variáveis)

Um exemplo seria uma tabela que gravasse dados por cliente. Para cada cliente teriamos nome, endereço, cidade, estado, data de nascimento, estado civil, renda.

O tensor a seguir tem 2 eixos, seu formato é (2,4) e ele armazena 2*4=8 números do tipo float32. Este tensor é chamado de "matriz"

```python
import numpy as np
a = np.array([[ 1, 2, 3, 4],
              [10,20,30,40]])
a, a.ndim, a.shape
```

#### Tensores

A séries temporais são geralmente tensores 3D. Neste caso temos (amostras, no.de instantes de tempo por amostra, variáveis).

Um exemplo seria um conjunto de paises (Argentina, Brasil, Chile, Paraguai, Uruguai). Para cada país seriam registrados dados de 1991 até 2010 (20 valores). Para cada ano seriam registradas as variáveis: População, PIB, No.de Nascimentos, No.de Mortes, Expectativa de Vida. 

O tensor a seguir tem 3 eixos, seu formato é (2,3,4) e ele armazena 2 * 3 * 4=24 números do tipo float32. A partir de 3D (3 eixos) chamamos simplesmente de tensor.

```python
import numpy as np
a = np.array([ [ [  1,  2,  3,  4],
                 [ 10, 20, 30, 40],
                 [100,200,300,400] ],
               [ [  5,  6,  7,  8],
                 [ 50, 60, 70, 80],
                 [500,600,700,800] ] ])
a, a.ndim, a.shape
```

O elemento 200 está neste caso na posição: (0, 2, 1). Observe:

```python
a[0, 2, 1]
```

#### Tensores 4D

Aqui temos geralmente imagens. Neste caso temos: (amostra, canal, altura, largura).

Amostra seria o número da imagem. 
Canal seria a cor (vermelho, verde ou azul)
Altura x Largura seria um número de 0 a 255 que indicará a intensidade da cor em cada ponto.

Por exemplo, suponha que vermelho (red) é 0 canal 0, verde (green) é o canal 1 e azul (blue) é o canal 2. 

Ainda neste exemplo as imagens são quadrados de 28 x 28 pontos.

Sendo assim (10, 1, 20, 10) = 150 significa o seguinte:

*   Imagem = 10a imagem
*   Cor 1 = Verde
*   Altura = 20 (ponto na vigésima linha)
*   Coluna = 10 (ponto na décima coluna)
*   Intensidade do verde = 150

No caso acima temos que: na 10a imagem, no canal de cor verde, no ponto na altura 20 e coluna 10, a intensidade do verde é 150.

#### Tensores 5D

Neste caso temos um conjunto de videos.

A codificação é feita com: (Amostra, Quadro, Canal, Altura, Largura)

Sendo assim (10, 1100, 1, 20, 12) = 150 quer dizer:

Para o filme de número 10, no quadro 1100, no canal 1 (cor verde por exemplo), no pixel na altura 20 e coluna 12, a intensidade do verde é 150.

# Deep Learning Utilizando a Biblioteca Keras

**Importação do conjunto de dados**

A célula a seguir importa o dataset mnist, o qual é um conjunto de imagens de números escritos a mão livre e digitalizados.

```python
from keras.datasets import mnist
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()
train_images.shape, test_images.shape, train_labels.shape, test_labels.shape
```

```python
import numpy as np
np.min(train_images), np.max(train_images), np.min(train_labels), np.max(train_labels)
```

**Importação da biblioteca gráfica e visualização de uma imagem**

```python
digit = train_images[4]
import matplotlib.pyplot as plt
plt.imshow(digit, cmap=plt.cm.binary)
plt.show()
```

**Importação da biblioteca gráfica e visualização de um grupo de imagens**

```python
from matplotlib import pyplot as plt
%matplotlib inline

fig = plt.figure(figsize=(10,4)) #Tamanho total da área de gráficos

for i in range(9):
  plt.subplot(3,3,i+1) # Qual o subplot em um grid 3x3
  plt.tight_layout() # Para permitir espaçamento ótimo entre os subplots
  
  #Desenho em tom de cinza de uma imagem (28x28) em train_images
  plt.imshow(train_images[i], cmap='gray', interpolation='none') 
  plt.title("Digit: {}".format(train_labels[i]))
  plt.xticks()
  plt.yticks()

fig
plt.show()
```

```python
from keras import models
from keras import layers
```

```python
network = models.Sequential()
```

```python
network.add(layers.Dense(512, activation='relu', input_shape=(28*28,)))
network.add(layers.Dense(10, activation='softmax'))
```

```python
network.compile(optimizer='rmsprop',
                loss='categorical_crossentropy',
                metrics=['accuracy'])
```

```python
from keras.utils import to_categorical
```

```python
train_labels2 = to_categorical(train_labels)
test_labels2 = to_categorical(test_labels)
train_labels[0], train_labels.shape, train_labels2[0], train_labels2.shape
```

```python
train_images2 = train_images.reshape(60000,28*28)
test_images2 = test_images.reshape(10000, 28*28)
train_images2.shape, test_images2.shape
```

**Normalização dos dados de entrada**

Vamos por último antes do treino normalizar os valores de entrada na faixa 0 a 1 

Valor normalizado = (Valor Original - Menor Valor)/(Maior Valor - Menor Valor)

Neste caso, como Menor Valor = 0 e Maior Valor = 255, normalizar significa dividir por 255

Apesar de não ser necessário normalizar os valores de saida, vamos criar um outro conjunto de índice 3 para manter o padrão com os valores de entrada

```python
train_images3 = train_images2 / 255
test_images3 = test_images2 / 255

train_labels3 = train_labels2
test_labels3 = test_labels2
```

```python
#train_images2 tem valores não normalizados
network.fit(train_images2, train_labels2, epochs=5, batch_size=128)
```

```python
test_loss, test_acc = network.evaluate(test_images2, test_labels2)
test_loss, test_acc
```

```python
#train_images3 tem valores normalizados
network.fit(train_images3, train_labels3, epochs=5, batch_size=128)
```

```python
test_loss, test_acc = network.evaluate(test_images3, test_labels3)
test_loss, test_acc
```
